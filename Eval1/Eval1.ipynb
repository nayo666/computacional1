```python
df1 = pd.DataFrame(data=df)
```

```python
df1.describe()
```

```python
import statsmodels.api as sm
lowess = sm.nonparametric.lowess
```

```python
low = lowess(df1[u'J-D'], df1[u'Year'], frac=0.3333)
```

```python
low1=low[:,0]
low2=low[:,1]
```

```python
mplt.plot(low1,low2)
```

```python
mplt.plot(df1[u'Year'],df1[u'J-D'])
```

```python
import pylab as mplt


mplt.plot(df1[u'Year'],df1[u'J-D'], c='black',label='Temperatura anual promedio')
mplt.plot(low1,low2, c='red',label='Temperatura suavizada con funcion lowess')
mplt.grid(True)
mplt.xlabel('Ciclo solar')
mplt.ylabel('Temperatura (C)')
mplt.title('Temperaturas promedio anuales (real y suavizada (lowess)')
mplt.legend(fancybox=True, shadow=True)
mplt.grid(True)

```
