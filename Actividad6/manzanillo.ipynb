```python
import pandas as pd

import numpy as np

import matplotlib as plt

%matplotlib inline
```

```python
df = pd.read_csv ("/home/noxd/computacional/a6/mnz.csv")
```

```python
#cambiar formato de tiempo
from datetime import datetime
df['date']= df.apply(lambda x:datetime.strptime("{0} {1} {2} {3}".format(x[u'anio'],x[u'mes'], x[u'dia'], x[u'hora(utc)']), "%Y %m %d %H"),axis=1)
```

```python
#eliminar variables extras
df_short=df.drop(df.columns[[0,1,2,3]],axis=1)
```

```python
df_short['altura(mm)']= df_short['altura(mm)'].astype(float).interpolate(method='spline', order=2)
```

```python
#indicador de tiempo
df_short['month']=pd.DatetimeIndex(df_short[u'date']).month
```

```python
df_dic=df_short.loc[df_short[u'month'] == 12]
```

```python


import pandas as pd
time_range = pd.date_range('2016-12-01 00:00:00','2016-12-31 23:00:00', freq='H')
# podemos aun seleccionar un rango de días
 
hours = time_range.hour
days = time_range.day
# Podemos calcular el número de horas transcurridas desde la fecha inicial
time_since_beginning = (days-1)*24+hours
time_since_beginning

```

```python
df_dic['hour']=time_since_beginning
```

```python
import matplotlib.pyplot as mplt
fig = mplt.gcf()
fig.set_size_inches(13.5, 7.5)


import numpy as np
from scipy.fftpack import rfft
# Number of samplepoints
N = 744
# sample spacing
T = 1.0 / 372
x = df_dic[u'hour']
#eliminando la suma
y = df_dic[u'altura(mm)']
yf = rfft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)
import matplotlib.pyplot as plt
plt.plot(xf, 2.0/N * np.abs(yf[0:N/2]))
plt.xlim(-5,80)
plt.ylim(0,800)
plt.grid()
plt.show()

```
