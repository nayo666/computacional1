```python
df_short['month']=pd.DatetimeIndex(df_short[u'Date Time']).month
```

```python
df_dic=df_short.loc[df_short[u'month'] == 12]
```

```python
df_dic
```

```python
# rango de días
time_range = pd.date_range('2016-12-01 00:00','2016-12-31 23:54', freq='6T')


#rangos de tiempo
mins = time_range.minute
hours = time_range.hour
days = time_range.day


#calcular el numero de cada 6 minutos
time_serie =  (days-1)*24*60+ hours*60 + mins
time_serie
```

```python
df_dic['6mins']=time_serie
```

```python
del df_dic[u'month']
```

```python
del df_dic[u'Date Time']
```

```python
df_dic
```

```python
import matplotlib.pyplot as mplt
fig = mplt.gcf()
fig.set_size_inches(13.5, 7.5)



import numpy as np
from scipy.fftpack import rfft
# Number of samplepoints
N = 7440
# sample spacing
T = 1.0 / 372
x = df_dic[u'6mins']
#eliminando la suma
y = df_dic[u' Water Level']
yf = rfft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)
import matplotlib.pyplot as plt
plt.plot(xf, 2.0/N * np.abs(yf[0:N/2]))
plt.xlim(-5,80)
plt.ylim(0,1.5)
plt.grid()
plt.show()

```
