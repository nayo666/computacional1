```python
import matplotlib.pyplot as mplt
fig = mplt.gcf()
fig.set_size_inches(13.5, 7.5)

df[u'manchas'] = df[u'manchas'].astype(float).interpolate(method='spline', order=2)

import numpy as np
from scipy.fftpack import rfft
# Number of samplepoints
N = 3212
# sample spacing
T = 1.0/132
x = df[u'añodec']
#eliminando la suma
y = df[u'manchas']
yf = rfft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)
import matplotlib.pyplot as plt
plt.plot(xf, 2.0/N * np.abs(yf[0:N/2]))
plt.xlim(-5,135)
plt.ylim(0,0.1)
plt.grid()
plt.show()
```
